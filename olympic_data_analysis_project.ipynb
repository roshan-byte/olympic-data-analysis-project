## Asking and Answering Questions

extracting some information by asking some questions and visualising the data using plotting function


# Q1. Top Five players who has won most number Gold Medals of medals? 

```python
plt.figure(figsize=(14,7))

athlete_medals=df[["Athlete","Medal"]]
athlete_gold=athlete_medals[athlete_medals["Medal"]=="Gold"].groupby("Athlete")

grouped=athlete_gold.count().sort_values("Medal",ascending=False).head(5)

name_of_athlete=grouped.index

no_of_gold_medal_win=grouped.Medal

plt.bar(name_of_athlete,no_of_gold_medal_win,color=['r','g','orange','k','b'],ec="k")

plt.xlabel("top five athlete who won maximum no of gold medals",fontsize=20)
plt.ylabel("no of gold medals won by the athlete",fontsize=20)

plt.title("top five athlete who won maximum no of gold in summer olympic ",fontsize=20)


#add label to each bar 

for i in range(len(name_of_athlete)):
    plt.text(i,no_of_gold_medal_win[i],no_of_gold_medal_win[i],ha="center",va="bottom",fontsize=12)
```

### Michael PHELPS who won the most number of Gold medals in swimming so far

# Q2: Top Five Sports  having most female Gold Medalists? 

```python
plt.figure(figsize=(14,7))
sport_gender=df[["Sport","Gender","Medal"]]

# sport_gender
female_sport=sport_gender[sport_gender["Gender"]=="Women"] 
female_sport_gold=female_sport[female_sport["Medal"]=="Gold"].groupby("Sport").count()


grouped=female_sport_gold.sort_values("Gender",ascending=False).head(5)

female_sport=grouped.index

no_of_female=grouped.Gender

plt.bar(female_sport,no_of_female,color=['r','g','orange','k','b'],ec="k")

plt.xlabel("top five sport having most female gold medalist",fontsize=20)
plt.ylabel("no of gold medals",fontsize=20)
plt.title("plot of top five sports having most female gold medalist",fontsize=22)

#labels for each bars

for i in range(len(female_sport)):
    plt.text(i,no_of_female[i],no_of_female[i],ha="center",va="bottom")
```

### Aquatics and Atheletics are the most populer games and also it is popular amongs female also

# Q3: Top Five player has won most number of medals? 

```python
plt.figure(figsize=(14,7))

#taking the field athlete and medal
athlete_medals=df[["Athlete","Medal"]].groupby("Athlete")
#groupby the data with athlete and count the medals win

grouped=athlete_medals.count().sort_values("Medal",ascending=False).head(5)

name_of_athlete=grouped.index

no_of_medal_win=grouped.Medal

plt.bar(name_of_athlete,no_of_medal_win,color=['r','g','orange','k','b'],ec="k")

plt.xlabel("top five athlete who won maximum medals",fontsize=20)
plt.ylabel("no of medals won by the athlete",fontsize=20)

plt.title("top five athlete who won maximum no of medals in summer olympic ",fontsize=20)


#add label to each bar 

for i in range(len(name_of_athlete)):
    plt.text(i,no_of_medal_win[i],no_of_medal_win[i],ha="center",va="bottom",fontsize=12)
```

### Michael Phelps is the only person who won maximum number of medals and also haivng maximum number of medals so far

# Q4 Top five sport  having most number of Gold Medals so far? 

```python
plt.figure(figsize=(14,7))

#list data where medal is gold medals
gold_medal=df[df['Medal']=="Gold"]

#taking two fields (i.e sport and Medals) and group them according to Sport
sport_gold_medal=gold_medal[["Sport","Medal"]].groupby("Sport")

#finding the count of gold medals in each sport
grouped=sport_gold_medal.count().sort_values("Medal",ascending=False)

#taking the top five sports having maximum no. of gold medals

sport=grouped.head(5).index

no_of_gold=grouped.head(5).Medal

plt.bar(sport,no_of_gold,color=['r','g','orange','k','b'],ec="k")

plt.xlabel("top five sports having maximum no. of gold medals()",fontsize=20)
plt.ylabel("no. of gold medals ",fontsize=20)

plt.title("List of Sports having maximum number of Gold medals so far ",fontsize=20)


#labels for each bars

for i in range(len(sport)):
    plt.text(i,no_of_gold[i],no_of_gold[i],ha="center",va="bottom")
```

### Aquatics Athletics are the most popular games and also most number of gold medals won in this game only
